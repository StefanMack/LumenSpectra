## Wieso sind optische Sensoren empfindlich gegenüber Umgebungslicht?
Alle optischen Sensoren können durch Umgebungslicht - der Fachbegriff dafür ist "Fremdlicht" - gestört werden. Je nach Sensorauslegung können schon wenige Hundert Lux zum Ausfall eines Sensors führen. Sensoren für den Außenbereich müssen Fremdlichtfestigkeiten von >100 klx aufweisen.
Die meisten optischen Sensoren senden eigenes Licht ("Nutzlicht") aus, das auf das anzutastende Objekt fällt. Das Objekt reflektiert einen Teil des Nutzlichts zurück zum Sensor.
Je nach Sensorverfahren d.h. je nachdem welche Objekteigenschaft gemessen werden soll, wie etwa der Objektabstand, die Objektfarbe oder die Objektpräsenz, wird die Lichtlaufzeit, das Lichtspektrum oder die Lichtintensität/-verteilung oder ein anderer Lichtparameter vom Sensor ausgewertet.  
  
Auf die Empfängeroptik des Sensors fällt jedoch nicht nur dieses Nutzlicht sondern zusätzlich auch Fremdlicht. Die Quelle des Fremdlichts ist meistens das Sonnenlicht oder in geschlossenen Räumen die Allgemeinbeleuchtung wie z.B. Halogen- oder LED-Leuchten.  
  
Nahezu alle optische Sensoren verwenden intensitätsmoduliertes Nutzlicht. Eine Infrarot-Fernbedienung puls z.B. das Nutzlicht mit einer Frequenz von 38 kHz. Da das Fremdlicht meistens Gleichlicht ist, kann so im Empfänger des Sensors der Fremdlichtanteil nach der Photodiode per Hochpassfilter  elektronisch unterdrückt werden. Dies funktioniert aber nur solange wie das Fremdlicht die Photodiode nicht übersteuert. Außerdem nimmt in jedem Fall das Rauschen des Nutzsignals durch das Fremdlicht zu und damit die Messgenauigkeit des Sensors ab.  
Hinzu kommt, dass die Allgemeinbeleuchtung ebenfalls oft intensitätsmoduliert ist: LEDs werden oft über Pulsweitenmodulation gedimmt, Leuchtstoffröhren werden mit Wechselspannung betrieben.  

Wenn es die Applikation und die Sensorkosten zulassen, wird möglichst Nutzlicht in dem Bereich verwendet, in dem das Spektrum des Fremdlichts eine geringe spektrale Leistungsdichte hat: Daher versucht man möglichst Infrarotlicht zu verwenden - je langwelliger destso besser, jedoch auch desto teurer.  

Weiter verwenden alle optischen Sensoren empfangsseitig optische Bandpassfilter, deren Durchlassbereich möglichst nur dem Nutzlichtspektrum entspricht. Das Fremdlicht wird hier spektral weitgehend ausgeblendet. Jedoch kommt innerhalb des Durchlassbereichs immer noch Fremdlicht zum Empfänger des Sensors.  
Daher verwendet man gerne schmalbandiges Laserlicht und passt man den Durchlassbereich des optischen Bandpassfilters so eng wie möglich daran an. Dies ist aber bei kostensensitiven Sensoren wie einfachen Lichttastern oder Lichtgittern nicht wirtschaftlich, weshalb hier meistens nur monochromatische LEDs als Nutzlichtquellen und nur eingefärbte Kunststoffscheiben als "Bandpassfilter" Verwendung finden.  
  
Bis auf wenige Ausnahmen im Militär- oder Weltraumbereich sind optische Sensoren trotzdem in der Praxis fremdlichtempfindlich. Denn Präzisionsoptiken wie genau gefertigte Linsen und Filter sind sehr kostspielig.  

## Wieso wird die Fremdlichtfestigkeit in Lux (lx) angegeben?
Damit ein Anwender abschätzen kann, ob die Verfügbarkeit eines optischen Sensors trotz Fremdlicht gegeben ist, muss die Fremdlichtfestigkeit des Sensors in irgend einer Weise spezifiziert sein.  
Idealerweise würde der Sensorhersteller hierfür die Nutzlichtwellenlänge und den Durchlassbereich des Sensorempfängers sowie die Modulation des Nutzlichts angeben. Dann könnte die Fremdlichtfestigkeit als maximale Bestrahlungsstärke für Gleichlicht bzw. moduliertes Licht spezifiziert werden.  
  
Eine solche Sensor-Spezifikation ist jedoch aus mehreren Gründen nicht praktikabel:  
1. Der Sensorhersteller möchte aus Gründen des IP-Schutzes nicht diese optischen Eigenschaften des Sensors veröffentlichen.
2. Der Anwender hat selten die Möglichkeit den Fremdlichtpegel spektral- und zeitaufgelöst zu bestimmen.  
  
Deshalb wird die Fremdlichtfestigkeit von Sensoren allgemein als maximale Beleuchtungsstärke in Lux auf dem Sensorempfänger spezifiziert.   
Zugegebenermaßen ist eine solche Spezifikation nicht sonderlich wasserdicht, denn sie berücksichtigt weder das Fremdlichtspektrum, eine Richtung des einfallenden Fremdlichts, noch eine mögliche Modulation des Fremdlichts.  
  
In Normen zu bestimmten Sensoren ist jedoch zur Bestimmung der Fremdlichtfestigkeit der Messaufbau für die zulässige Beleuchtungsstärke näher spezifiziert. Z.B. wird ein Halogenscheinwerfer mit einer bestimmten elektrischen Leistung und Reflektorgröße vorgeschrieben, der entlang der Sichtlinie des Sensors in den Sensor hinein strahlt.  

## Fremdlicht-Bestrahlungsstärke aus der Beleuchtungsstärke abschätzen
Ist das Spektrum des Fremdlichts, die Nutzlichtwellenlänge, sowie die Breite des Emfpänger-Bandpassfilters bekannt, so kann aus der Beleuchtungsstärke die Fremdlicht-Bestrahlungsstärke auf der Photodiode des Sensors (also nach dem Bandpassfilter) wie folgt abgeschätzt werden.  

Nachfolgend wird diese Berechung für Halogenlicht und Sonnenlicht als Fremdlichtquelle durchgeführt.  
Die Spektraldaten für Halogenlicht befinden sich in der Textdatei ``Halogen.txt``, die für Sonnenlicht in der Textdatei ``Sonne.txt``. In den Textdateien gibt die erste Spalte die Wellenlänge und die zweite Spalte die spektrale Bestrahlungsstärke (Leistung (a.u.) pro m² und nm) an.  
Die Textdatei ``PhotopAutenempf.txt`` gibt die normierte Augenempfindlichkeitskurve eines menschlichen Auges für das Hellempfinden an. Diese Kurve hat ihr Maximum bei einer Wellenlänge von 555 nm. An diesem Punkt entspricht einer Bestrahlungsstärke von 1 W/m² einer Beleuchtungsstärke von 683 lx.  

Zunächst werden die nötigen Bibliotheken sowie die Spektraldaten von Auge, Fremdlicht und Bandpassfilter geladen:

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm # fuer Darstellung des Nutzlichtbereichs

wv_sens_light = 650 # Wellenlaenge Nutzlicht, Defaultwert 650 nm
wv_fwhm_sens_light =15 # Halbwertsbr. Durchlassber. Bandpass. Nutzlicht, Defaultwert 15 nm

wv_hal, spec_hal = np.loadtxt('Halogen.txt', unpack=True)
wv_sun, spec_sun = np.loadtxt('Sonne.txt', unpack=True)
wv_sco, spec_sco = np.loadtxt('ScotopAugenempf.txt', unpack=True) # Dunkelempfinden
wv_pho, spec_pho = np.loadtxt('PhotopAugenempf.txt', unpack=True) # Hellempfinden

# Spektren normieren auf 1 fuer 555 nm
spec_hal_n = spec_hal/spec_hal[np.where(wv_hal==555)[0][0]]
spec_sun_n = spec_sun/spec_sun[np.where(wv_sun==555)[0][0]]
```

Die Spektren für Halogenlicht und Sonnenlicht wurden oben auf 555 nm normiert. Für die tatsächliche Bestrahlungsstärke dieser Fremdlichtquellen muss nun noch die spektrale Bestrahlungsstärke für eine Wellenlänge von 555 nm in de Einheit W/nm/m² angegeben werden:

```python
irrad_sp_hal = 0.7 # Defaultwert 0,8 W/nm/m² (ergibt ca. 50 klx)
irrad_sp_sun = 1.5 # Defaultwert 1,5 W/nm/m² (ergibt ca. 100 klx)
print('Spektrale Bestrahlungsstärke: Halogenleuchte {} W/nm/m², Sonnenlicht {} W/nm/m²'
      .format(irrad_sp_hal,irrad_sp_sun))
```

Aus den obigen Eingabedaten für den Fremdlichtpegel wird nun die Beleuchtungsstärke der beiden Fremdlichtquellen in Lux berechnet.  
Dafür wird die spektrale Augenempfindlichkeitskurve mit dem Fremdlichtspektrum gefaltet.

```python
# Arrayindex berechnen fuer Start-/Endwellenlaenge Augenempfindlichkeit
wv_start_hal = np.where(wv_hal==wv_pho[0])[0][0]
wv_end_hal = np.where(wv_hal==wv_pho[-1])[0][0]
wv_start_sun = np.where(wv_sun==wv_pho[0])[0][0]
wv_end_sun = np.where(wv_sun==wv_pho[-1])[0][0]

# Augenempfindlichkeitskurve und Spektren falten
# ACHTUNG: Beide Kurven muessen gleiche Schrittweite bei Wellenlaengen haben!
d_wv = wv_pho[1] - wv_pho[0] # Schrittweite Wellenlaenge in nm
# Beleuchtungsstaerke = Skalarprodukt * Schrittweite * 
illum_hal = np.dot(spec_pho,spec_hal_n[wv_start_hal:wv_end_hal+1]) * d_wv * irrad_sp_hal * 683
illum_sun = np.dot(spec_pho,spec_sun_n[wv_start_sun:wv_end_sun+1]) * d_wv * irrad_sp_sun * 683
print('Beleuchtungsstärke: Halogenleuchte {:6.0f} lux, Sonnenlicht {:6.0f} lux'.format(illum_hal,illum_sun))
```

Nachfolgend wird schließlich die Bestrahlungsstärke auf dem Sensorempfänger durch die beiden Fremdlichtquellen berechnet. Dabei wird das Spektrum des Bandpassfilters näherungsweise als gaußförmig angenommen. Die Halbwertsbreite dieser Gaußkurve entspricht dem FWHM (Full Width at Half Maximum) des Bandpassfilters.

```python
irrad_sen_hal = spec_hal_n[np.where(wv_hal == wv_sens_light)[0][0]] * irrad_sp_hal * wv_fwhm_sens_light
irrad_sen_sun = spec_sun_n[np.where(wv_sun == wv_sens_light)[0][0]] * irrad_sp_sun * wv_fwhm_sens_light

print('Nutzlichtbereich: {} nm +/-{} nm (FWHM)'.format(wv_sens_light,wv_fwhm_sens_light))
print('Bestrahlungsstärke Nutzlichtbereich: Halogenleuchte {:6.3f} W/m², Sonnenlicht {:6.3} W/m²'.format(irrad_sen_hal,irrad_sen_sun))
```

Zur Visualisierung der verschiedenen Spektren werden diese schließlich grafisch dargestellt.

```python
# Spektrum Nutzlichtbereich
wv = np.arange(300,900,0.5) # Wellenlaengenwerte Nutzlicht
# Skalierung PDF (Gausfunktion fuer FWHM) siehe 
# docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html
scale_gauss = 1/1.177*wv_fwhm_sens_light/2 
spec_sens = norm.pdf(wv,wv_sens_light,scale_gauss) * scale_gauss * 2.507 # Durchlass Bandpass Sensor


fig, ax = plt.subplots() 
ax.plot(wv_hal,spec_hal_n,label='Halogenleuchte',linewidth=3.0,alpha=0.7)
ax.plot(wv_sun,spec_sun_n,label='Sonnenlicht',linewidth=3.0,alpha=0.7)
ax.plot(wv_pho,spec_pho,label='Augenempf. Hell',linewidth=3.0,alpha=0.7)
ax.plot(wv,spec_sens,label='Bandpass Sensor',linewidth=3.0,alpha=0.7)

ax.grid(linestyle=':')
ax.set_xlabel('Wellenlänge (nm)',fontsize=14)
ax.set_ylabel('Leistungsdichte/Empfindlichkeit normiert',fontsize=14)
ax.legend() # Legende mit Werten aus ax.plot(label='')
ax.legend(fontsize=12)
plt.tight_layout()
fig.savefig('SpektrenLichtquellen.png')
```

## Fazit
Die Default-Werte in den Berechnungen oben für 100 klx Sonnenlicht (Sonnenschein im Sommer) zeigen, dass für rotes Lasernutzlicht auf die Empfängerdiode trotz Bandpassfilter immer noch eine Bestrahlungsstärke von etwa 19,6 W/m²  fallen. Bei einer üblichen Empfängeraperturgröße von 100 mm² entspricht dies immer noch einer optischen Leistung von ca. 2 mW, welche auf die Photodiode fällt.  
Nutzlichtleistungen hingegen bewegen sich im µW Bereich und darunter. Bei Lidar-Sensoren sind es meist nur wenige nW. Dies verdeutlicht welche hohen technischen (optische wie elektronische) Anforderungen an einen im Außenbereich fremdlichtfesten Sensor gestellt werden.
